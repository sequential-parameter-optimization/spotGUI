```python
import pandas as pd
from river import datasets
from spotRiver.utils.data_conversion import convert_to_df
dataset = datasets.Bikes()
target_column = "y"
df = convert_to_df(dataset, target_column)
df.head()
```

```python
# Select clouds', 'humidity', 'pressure', 'temperature', 'wind' as features and y as target and save it to a new dataframe
# convert y to float
df = df[['clouds', 'humidity', 'pressure', 'temperature', 'wind', 'y']]
df['y'] = df['y'].astype(float)
df.head()
```

```python
# save the df to a csv file
df.to_csv('bikes_numerical_features.csv', index=False)
```
